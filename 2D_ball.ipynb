```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plotting the grid and unit circle for all three cases
def plot_2D_ball(p):

    # Defining the graph space
    x = np.linspace(-5, 5, 100)
    y = np.linspace(-5, 5, 100)
    X, Y = np.meshgrid(x, y)

    # Computation criteria for the norm of each point, i.e. p = 1, 2, ..., infinity
    if p == np.inf:
        N = np.maximum(np.abs(X), np.abs(Y))
    elif p == 1 or p == 2:
        N = (np.abs(X) ** p + np.abs(Y) ** p) ** (1/p)
    else:
        raise ValueError("This is an invalid value for p; p = 1, 2, or infinity. Please input the value of p once again")

    # Plotting conditions
    plt.figure(figsize=(8, 6))
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title(f'Unit Circle with p-norm (p={p})')
    plt.contourf(X, Y, N, levels=20, cmap='viridis')
    plt.colorbar(label='Norm')
    plt.grid(True)
    plt.show()
```

```python
# Values of p for different norms
p_values = [1, 2, np.inf]

# Plot unit circles for different p-values
for p in p_values:
    plot_2D_ball(p)
```
