```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Randomly Generated 2D Data with Size = 200 for the Linear Case
np.random.seed(250)
X1 = np.random.normal(15, 3, 200)
Y1 = 5 * X1 + 2 + np.random.normal(20, 4, 200)
```

```python
plt.scatter(X1,Y1)
```

```python
# Randomly Generated 2D Data with Size = 200 for the Quadratic Case
np.random.seed(250)
X2 = np.random.normal(30, 12, 200)
Y2 = X2 ** 2 + 2 * X2 + 1 + np.random.normal(35, 14, 200)
```

```python
plt.scatter(X2,Y2)
```

```python
# Randomly Generated 2D Data with Size = 200 for the Random Scatter Case
np.random.seed(250)
X3 = np.random.rand(200) * 10
Y3 = np.random.rand(200) * 5 + 5 * X3 + 2 + np.random.normal(20, 4, 200)
```

```python
plt.scatter(X3,Y3)
```

```python
# Define Mean squared error/loss function for the Linear Case
def loss_function(Y1, Y1_pred):
    MSE1 = np.mean((Y1 - Y1_pred) ** 2)
    return MSE1
```

```python
# Define Mean squared error/loss function for the Quadratic Case
def loss_function(Y2, Y2_pred):
    MSE2 = np.mean((Y2 - Y2_pred) ** 2)
    return MSE2
```

```python
# Define Mean squared error/loss function for the Random Scatter Case
def loss_function(Y3, Y3_pred):
    MSE3 = np.mean((Y3 - Y3_pred) ** 2)
    return MSE3
```

```python
# Gradient Descent function for the Linear Case
def gradient_descent(X1, Y1, learning_rate = 0.0001, iterations = 1000):
    n1 = len(Y1)
    m1 = 0
    c1 = 0
    for i in range(iterations):
        Y1_pred = m1 * X1 + c1
        D1 = Y1 - Y1_pred
        m1_gradient = (-2/n1) * np.sum(D1 * X1)
        c1_gradient = (-2/n1) * np.sum(D1)
        m1 -= learning_rate * m1_gradient
        c1 -= learning_rate * c1_gradient
    return m1,c1
# Linear regression using the gradient descent function for the Linear Case
m1, c1 = gradient_descent(X1, Y1, learning_rate = 0.0001, iterations = 1000)
```

```python
# Gradient Descent function for the Quadratic Case
def gradient_descent(X2, Y2, learning_rate = 0.0001, iterations = 1000):
    n2 = len(Y2)
    m2 = 0
    c2 = 0
    for i in range(iterations):
        Y2_pred = m2 * X2 + c2
        D2 = Y2 - Y2_pred
        m2_gradient = (-2/n2) * np.sum(D2 * X2)
        c2_gradient = (-2/n2) * np.sum(D2)
        m2 -= learning_rate * m2_gradient
        c2 -= learning_rate * c2_gradient
    return m2,c2
# Linear regression using the gradient descent function for the Linear Case
m2, c2 = gradient_descent(X2, Y2, learning_rate = 0.0001, iterations = 1000)
```

```python
# Gradient Descent function for the Random Scatter Case
def gradient_descent(X3, Y3, learning_rate = 0.0001, iterations = 1000):
    n3 = len(Y3)
    m3 = 0
    c3 = 0
    for i in range(iterations):
        Y3_pred = m3 * X3 + c3
        D3 = Y3 - Y3_pred
        m3_gradient = (-2/n3) * np.sum(D3 * X3)
        c3_gradient = (-2/n3) * np.sum(D3)
        m3 -= learning_rate * m3_gradient
        c3 -= learning_rate * c3_gradient
    return m3,c3
# Linear regression using the gradient descent function for the Linear Case
m3, c3 = gradient_descent(X3, Y3, learning_rate = 0.0001, iterations = 1000)
```

```python
# Line of Best fit function for the Linear Case
Y1_pred = m1 * X1 + c1
```

```python
# Line of Best fit function for the Quadratic Case
Y2_pred = m2 * X2 + c2
```

```python
# Line of Best fit function for the Random Scatter Case
Y3_pred = m3 * X3 + c3
```

```python
# Plot the results and line of best fit for Linear Case
plt.scatter(X1, Y1, label='Data points')
plt.plot(X1, m1 * X1 + c1, color='red', label='Line of Best Fit')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Best Linear Fit using Gradient Descent for Linear Case')
plt.legend()
```

```python
# Plot the results and line of best fit for Quadratic Case
plt.scatter(X2, Y2, label='Data points')
plt.plot(X2, m2 * X2 + c2, color='red', label='Line of Best Fit')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Best Linear Fit using Gradient Descent for Quadratic Case')
plt.legend()
```

```python
# Plot the results and line of best fit for Random Scatter Case
plt.scatter(X3, Y3, label='Data points')
plt.plot(X3, m3 * X3 + c3, color='red', label='Line of Best Fit')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Best Linear Fit using Gradient Descent for Random Scatter Case')
plt.legend()
```

```python
# Final parameters for Linear Case
MSE1 = loss_function(Y1, Y1_pred)
print("The Final Parameters are:")
print("The Intercept = ", c1)
print("The Slope = ", m1)
print("The mean squared error = ", MSE1)
```

```python
# Final parameters for Linear Case
MSE2 = loss_function(Y2, Y2_pred)
print("The Final Parameters are:")
print("The Intercept = ", c2)
print("The Slope = ", m2)
print("The mean squared error = ", MSE2)
```

```python
# Final parameters for Linear Case
MSE3 = loss_function(Y3, Y3_pred)
print("The Final Parameters are:")
print("The Intercept = ", c3)
print("The Slope = ", m3)
print("The mean squared error = ", MSE3)
```
